```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep

# Initialize WebDriver (Make sure to replace with your Chrome driver path)
chrome_driver_path = "C:\\Users\\pc\\OneDrive\\Documents\\chromedriver-win64\\chromedriver.exe"
driver = webdriver.Chrome(chrome_driver_path)
url = "https://www.booking.com/searchresults.html?ss=Morocco&ssne=Morocco&ssne_untouched=Morocco&efdco=1&label=gen173nr-1FCAQoggJCDHNlYXJjaF9tYXJvY0gNWARojAGIAQGYAQ24ARfIAQzYAQHoAQH4AQOIAgGoAgO4Aq7T4rIGwAIB0gIkM2Q4ZjgzNGItNzBmYy00NDVkLWJhNTUtZTM5NmNlN2JhNGU52AIF4AIB&aid=304142&lang=en-us&sb=1&src_elem=sb&src=searchresults&dest_id=143&dest_type=country&checkin=2024-07-02&checkout=2024-07-03&group_adults=2&no_rooms=1&group_children=0"
driver.get(url)
sleep(600)
df = pd.DataFrame(columns=['hotel_name',
                'hotel_link',
                'address',
                'description',
                'facilities',
                'ville',
                'hotel_caracterstic',
                'price'])
print("weak up")
# Get the page source after waiting
src = driver.page_source

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(src, 'html.parser')
hotels=soup.find_all('div',{'class':'c066246e13 d8aec464ca'})
j=1
print(len(hotels))
for hotel in hotels:
    facilities = []
    hotel_caracterstic=[]
    print("----"*30)
    print('hotel number ',j)
    hotel_name=hotel.find('h3',{'class':'aab71f8e4e'}).text.strip().replace('Opens in new window', '').strip()
    hotel_link=hotel.find('h3',{'class':'aab71f8e4e'}).find('a',{'class':'a78ca197d0'}).get('href')
    price=hotel.find('span',{'class':'f6431b446c fbfd7c1165 e84eb96b1f'}).text.strip().replace('MAD', '').strip()
    ville=hotel.find('span',{'class':'aee5343fdb def9bc142a'}).text.strip()
    cara=hotel.find('div',{'class':'c59cd18527'}).find('h4',{'class':'abf093bdfe e8f7c070a7'}).text.strip()
    hotel_caracterstic.append(cara)
    li=hotel.find('div',{'class':'c59cd18527'}).find_all('li',{'class':'a6a38de85e'})
    for l in li:
        c=l.text.strip()
        hotel_caracterstic.append(c)
    print(ville)
    print(hotel_caracterstic)
    print(hotel_name)
    print(hotel_link)
    print(price)
    driver.get(hotel_link)
    sleep(6)
    # Get the page source after waiting
    src1 = driver.page_source
    # Parse the page source with BeautifulSoup
    soup1 = BeautifulSoup(src1, 'html.parser')
    address_element = soup1.find('span', {'class': 'hp_address_subtitle'})
    address = address_element.text.strip() if address_element else 'No address available'
    description_element = soup1.find('p', {'data-testid': 'property-description'})
    description = description_element.text.strip() if description_element else 'No description available'
    ul=soup1.find('ul', class_='c807d72881 d1a624a1cc e10711a42e')
    lis=ul.find_all('li', class_='a8b57ad3ff d50c412d31 fb9a5438f9 c7a5a1307a')
    print("li number",len(lis))
    i=1
    for li in lis:
        if i<=len(lis):
            facility = li.find('span', class_='a5a5a75131').text.strip()
            facilities.append(facility)
            print(i)
            print(facility)
            i+=1
    print(description)
    print(address)
    df = df.append({
                'hotel_name': hotel_name,
                'hotel_link': hotel_link,
                'address':address,
                'description':description,
                'facilities':facilities,
                'ville':ville,
                'hotel_caracterstic':hotel_caracterstic,
                'price':price
                
            }, ignore_index=True)
    j=j+1
    print("----"*30)
driver.quit()
```
