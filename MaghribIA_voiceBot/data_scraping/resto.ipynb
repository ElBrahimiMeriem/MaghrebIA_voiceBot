```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep

# Initialize WebDriver (Make sure to replace with your Chrome driver path)
chrome_driver_path = "C:\\Users\\pc\\OneDrive\\Documents\\chromedriver-win64\\chromedriver.exe"
driver = webdriver.Chrome(chrome_driver_path)
url = "https://www.bestrestaurantsmaroc.com/en/"
driver.get(url)
df = pd.DataFrame(columns=['city', 'cuisine', 'resto name', 'ville', 'capacity_details', 'description', 'ful_link', 'menu_items'])
base_url="https://www.bestrestaurantsmaroc.com"
sleep(6)
print("wake up")

# Get the page source after waiting
src = driver.page_source
base_url="https://www.bestrestaurantsmaroc.com/"
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(src, 'html.parser')
cities = []

for slide in soup.find_all('div', class_='swiper-slide'):
    city = slide.find('h3', class_='rs-cat-title').text
    link = slide.find('a', class_='rs-link')['href']
    cities.append({'city': city, 'link': link})
    full_link = base_url + link
    driver.get(full_link)
    sleep(6)
    src1 = driver.page_source
    soup1 = BeautifulSoup(src1, 'html.parser')
    restaurants = []

    for restaurant in soup1.find_all('div', class_='filterData'):
        title = restaurant.find('h3', class_='rs-cat-title rs-info').text
        link2 = restaurant.find('a', class_='rs-link')['href']
        cuisine = restaurant.find('div', class_='rs-cuisine rs-info').find('a').text
        ville = restaurant.find('div', class_='rs-city rs-info').text
        ful_link = base_url + link2
        driver.get(ful_link)
        sleep(6)
        src2 = driver.page_source
        soup2 = BeautifulSoup(src2, 'html.parser')
        description=''
        desc=soup2.find('div', class_='rs-description')
        if desc:
            description = desc.text

        capacity_details = {}
        for span in soup2.find('div', class_='rs-details').find_all('span', class_='br-bg-grey'):
            parts = span.text.split('\xa0:\xa0')
            if len(parts) == 2:
                capacity_type = parts[0].strip()
                capacity_value = parts[1].strip().replace('\xa0', ' ').strip()
                capacity_details[capacity_type] = capacity_value

        menu_sections = soup2.find_all('div', class_='rs-menu-section')
        menu_items = ''
        for section in menu_sections:
            section_title_tag = section.find('h4', class_='rs-item-title rs-title br-title')
            if section_title_tag:
                menu_items += section_title_tag.text.strip()+'\n'
                items = section.find_all('div', class_='rs-menu-item')
                for item in items:
                    item_name_tag = item.find('span', class_='rs-menu-desc')
                    if item_name_tag:
                        menu_items += item_name_tag.text.strip()+'\n'
                        

        df = df.append({
            'city': city,
            'cuisine': cuisine,
            'resto name': title,
            'ville': ville,
            'capacity_details': capacity_details,
            'description': description,
            'ful_link': ful_link,
            'menu_items': menu_items
        }, ignore_index=True)

        print(city)
        print(cuisine)
        print(title)
        print(ville)
        print(capacity_details)
        print(description)
        print(ful_link)
        print(menu_items)
        print("----" * 30)

driver.quit()
```
