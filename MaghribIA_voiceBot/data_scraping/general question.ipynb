```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep

# Initialize WebDriver (Make sure to replace with your Chrome driver path)
chrome_driver_path = "C:\\Users\\pc\\OneDrive\\Documents\\chromedriver-win64\\chromedriver.exe"
driver = webdriver.Chrome(chrome_driver_path)
df = pd.DataFrame(columns=['question', 'response'])

# Base URL for Lonely Planet Morocco attractions
url = "https://www.nomadicmatt.com/travel-guides/morocco-travel-guide/#costs"
driver.get(url)
sleep(5)  # Add a pause to ensure content loads properly
    
    # Récupération du contenu de la page avec BeautifulSoup
src = driver.page_source
soup = BeautifulSoup(src, 'html.parser')
guides=soup.find_all('div',{'class':'guide-block'})
for guide in guides[1:]:
    # Find the container that holds the links
    question = guide.find('h3').get_text(strip=True)

    # Find all the content after the h3 tag for the answer
    content = guide.find('h3').find_next_siblings()

    answer = ''
    for element in content:
        if element.name == 'h3':  # Stop if we reach another h3 tag
            break
        answer += element.get_text(strip=True) + '\n'

print("Question: ", question)
print("Answer: ", answer)
# Close the WebDriver
driver.quit()

```

```python
df = df.append({
                    'question': question,
                    'response': answer,
                    
                }, ignore_index=True)
```
