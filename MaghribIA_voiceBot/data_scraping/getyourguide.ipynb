```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep

# Initialize WebDriver (Make sure to replace with your Chrome driver path)
chrome_driver_path = "C:\\Users\\pc\\OneDrive\\Documents\\chromedriver-win64\\chromedriver.exe"
driver = webdriver.Chrome(chrome_driver_path)

# DataFrame initialization
df = pd.DataFrame(columns=['href', 'title', 'rating', 'reviews', 'price_low', 'price_per_person', 'highlights', 'full_description', 'includes', 'excludes', 'not_suitable_for', 'what_to_bring'])

bese = 'https://www.getyourguide.com/'
url = 'https://www.getyourguide.com/morocco-l169143/'
driver.get(url)
sleep(300)  # Add a pause to ensure content loads properly
print('weak up')   
# Skip the first "Show more" button and click subsequent ones

# Récupération du contenu de la page avec BeautifulSoup
#sleep(5)
src = driver.page_source
soup = BeautifulSoup(src, 'html.parser')
container = soup.find('section', {'class': 'grid-wrapper container'})
items = soup.find_all('div', {'class': 'vertical-activity-card__content-wrapper'})
print(len(items))

for item in items:
    try:
        href = item.find('a', class_='vertical-activity-card__container')['href']

        # Extracting title
        title = item.find('h3', class_='vertical-activity-card__title').get_text(strip=True)

        # Extracting image URL
        image = item.find('div', class_='vertical-activity-card__photo').find('img')['src']

        # Extracting description
        description_elements = item.find('div', class_='vertical-activity-card__body').find_all('span', class_='text-atom')
        description = ' • '.join([element.get_text(strip=True) for element in description_elements])

        # Extracting rating and number of reviews
        rating = item.find('div', class_='c-activity-rating__rating').get_text(strip=True)
        reviews = item.find('div', class_='c-activity-rating__label').get_text(strip=True)

        # Extracting prices
        price_low_element = item.find('div', class_='baseline-pricing__value baseline-pricing__value--low')
        price_low = price_low_element.find('span', class_='text-atom text-atom--body-compact-strong').get_text(strip=True) if price_low_element else "Not available"

        price_per_person_element = item.find('p', class_='text-atom--caption')
        price_per_person = price_per_person_element.get_text(strip=True) if price_per_person_element else "Not available"

        url1 = bese + href
        driver.get(url1)
        sleep(5)  # Add a pause to ensure content loads properly

        # Récupération du contenu de la page avec BeautifulSoup
        src1 = driver.page_source
        soup1 = BeautifulSoup(src1, 'html.parser')

        # Extracting details
        details = []
        terms = soup1.find_all('dt', class_='activity-key-details__term')
        descriptions = soup1.find_all('dd', class_='activity-key-details__description')

        for term, description in zip(terms, descriptions):
            details.append(f"{term.get_text(strip=True)}: {description.get_text(strip=True)}")

        # Creating a paragraph from the details
        paragraph = " ".join(details)

        # Extracting highlights
        highlights_section = soup1.find('section', {'id': 'highlights'})
        if highlights_section:
            highlights = highlights_section.find_all('li', class_='activity-highlights__list-item')
            highlights_text = [item.get_text(strip=True) for item in highlights]
        else:
            highlights_text = []

        # Extracting full description
        full_description_section = soup1.find('section', {'class': 'full-description'})
        if full_description_section:
            full_description = full_description_section.find('div', class_='toggle-content__content').get_text(strip=True)
        else:
            full_description = ""

        # Extracting includes and excludes
        includes_section = soup1.find('section', {'class': 'activity-includes'})
        if includes_section:
            includes = includes_section.find_all('ul', class_='activity-inclusions-inclusion')
            includes_text = [item.get_text(strip=True) for item in includes]
        else:
            includes_text = []

        excludes_section = soup1.find('section', {'class': 'activity-includes'})
        if excludes_section:
            excludes = excludes_section.find_all('ul', class_='activity-inclusions-exclusion')
            excludes_text = [item.get_text(strip=True) for item in excludes]
        else:
            excludes_text = []

        # Extracting not suitable for
        not_suitable_for_section = soup1.find('section', {'id': 'not-suitable-for'})
        if not_suitable_for_section:
            not_suitable_for = not_suitable_for_section.find_all('li')
            not_suitable_for_text = [item.get_text(strip=True) for item in not_suitable_for]
        else:
            not_suitable_for_text = []

        # Extracting what to bring
        important_info_section = soup1.find('section', {'id': 'section--important-information'})
        if important_info_section:
            what_to_bring_section = important_info_section.find('div', class_='activity-important-information__content')
            if what_to_bring_section:
                what_to_bring_title = what_to_bring_section.find('p', class_='activity-important-information__title').get_text(strip=True)
                what_to_bring_items = what_to_bring_section.find_all('li', class_='activity-important-information__item')
                what_to_bring_list = [item.get_text(strip=True) for item in what_to_bring_items]
            else:
                what_to_bring_title = ""
                what_to_bring_list = []
        else:
            what_to_bring_title = ""
            what_to_bring_list = []

        # Adding data to DataFrame
        df = df.append({
            'href': href,
            'title': title,
            'rating': rating,
            'reviews': reviews,
            'price_low': price_low.replace('د.م',''),
            'price_per_person': price_per_person,
            'About this activity': 'About this activity '+paragraph,
            'highlights': ' • '.join(highlights_text),
            'full_description': full_description,
            'includes': ' • '.join(includes_text),
            'excludes': ' • '.join(excludes_text),
            'not_suitable_for': ' • '.join(not_suitable_for_text),
            'what_to_bring': ' • '.join(what_to_bring_list)
        }, ignore_index=True)
    except Exception as e:
        print(f"Error processing item: {e}")

# Save DataFrame to a CSV file
df.to_csv('activities_morocco.csv', index=False)

driver.quit()
```
