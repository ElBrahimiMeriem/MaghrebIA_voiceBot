```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep

# Initialize WebDriver (Make sure to replace with your Chrome driver path)
chrome_driver_path = "C:\\Users\\pc\\OneDrive\\Documents\\chromedriver-win64\\chromedriver.exe"
driver = webdriver.Chrome(chrome_driver_path)
url = "https://www.memphistours.com/Morocco/Morocco-Travel-Guide/Morocco-Cities"
driver.get(url)
df = pd.DataFrame(columns=['question', 'response'])

sleep(5)
# Get the page source after waiting
src = driver.page_source

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(src, 'html.parser')
city_containers = soup.find_all('div', class_='home_special_offer')

# Extract information for each city
for city in city_containers:
    # Extract title
    title_tag = city.find('div', class_='home_special_offer_title').find('h2').find('a')
    title = title_tag.get_text(strip=True) if title_tag else 'N/A'
    urlbase = 'https://www.memphistours.com'
    # Extract URL
    link = title_tag['href'] if title_tag else 'N/A'
    full_link = urlbase + link
    driver.get(full_link)
    sleep(5)
    # Get the page source after waiting
    src2 = driver.page_source
    soup2 = BeautifulSoup(src2, 'html.parser')
    containers = soup2.find_all('div', {'class': 'program_box_data'})
    
    # Collect paragraphs without previous <h3> siblings
    p = ""
    for container in containers:
        for p_tag in container.find_all('p'):
            previous_siblings = p_tag.find_previous_siblings('h3')
            if not previous_siblings:
                p += p_tag.text.strip() + " "
    
    df = df.append({
        'question': title,
        'response': p.strip()
    }, ignore_index=True)
    
    print("Question:", title)
    print("Response:", p.strip())
    print("--------------------")
    
    # Collect <h3> tags as questions and their following <p> siblings as responses
    for container in containers:
        h3_tags = container.find_all('h3')
        for h3_tag in h3_tags:
            question = h3_tag.get_text(strip=True)
            response = ""
            next_element = h3_tag.find_next_sibling()
            while next_element and next_element.name != 'h3':
                if next_element.name == 'p':
                    response += next_element.get_text(strip=True) + " "
                next_element = next_element.find_next_sibling()
            
            df = df.append({
                'question': question,
                'response': response.strip()
            }, ignore_index=True)
            
            print("Question:", question)
            print("Response:", response.strip())
            print("--------------------")

# Close the WebDriver
driver.quit()

# Display or save the DataFrame if needed
print(df)
```
