```python
from urllib.parse import urlparse, urlunparse
def get_base_url(url):
    parsed_url = urlparse(url)
    base_url = urlunparse((parsed_url.scheme, parsed_url.netloc, parsed_url.path, '', '', ''))
    return base_url
```

```python
def create_prompt_response_pairs(row):
    hotel_name = row["hotel_name"]
    hotel_link = get_base_url(row["hotel_link"])
    address = row["address"]
    description = row["description"]
    facilities = ", ".join(ast.literal_eval(row["facilities"])) if isinstance(row["facilities"], str) else row["facilities"]
    ville = row["ville"]
    hotel_caracteristic = ", ".join(row["hotel_caracterstic"]) if isinstance(row["hotel_caracterstic"], list) else row["hotel_caracterstic"]
    price = row["price"]


    prompts = [
        f"{hotel_name} {ville}",
        f"hotel facilities: {facilities}",
        f"hotel in {ville}"
        
    ]

    response = (
    f"{hotel_name} in {ville}. "
    f"The hotel is located at {address}. It offers {facilities} and is priced at {price} for two adults. "
    f"Room characteristics {hotel_caracteristic}. "
    f"{description} "
    f"You can find more information at {hotel_link}."
    )


    return [(prompt, response) for prompt in prompts]

# Step 3: Apply Function to DataFrame and Generate Prompt-Response Pairs
all_pairs = []
for index, row in df.iterrows():
    pairs = create_prompt_response_pairs(row)
    all_pairs.extend(pairs)

# Step 4: Save Generated Pairs to JSON File
output_file = "hotel_prompt_response.json"
```

```python
df_pairs = pd.DataFrame(all_pairs, columns=["Prompt", "Response"])

# Display the DataFrame
print(df_pairs)
```
